## 유방암 진단기
---

### 데이터 세팅
---

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

cancer  = load_breast_cancer()

cancer_data = cancer.data
cancer_label = cancer.target

# 문제지 / 시험지 준비
x_train, x_test, y_train, y_test = train_test_split(cancer_data, 
                                                    cancer_label, 
                                                   test_size=0.2,
                                                   random_state=24)

print('x_train:', len(x_train), 'x_test:', len(x_test))
```

### 데이터 분석
---

```python
# 데이터에 담긴 정보
print('keys:', cancer.keys())

# 데이터와 라벨의 크기
print('data_shape:', cancer_data.shape)
print('label_shape:', cancer_label.shape)

# feature들의 이름
print('feature_names:', cancer.feature_names)

# 라벨의 이름
print('target_names:', cancer.target_names)

# 데이터셋에 대한 설명 
print(cancer.DESCR)
```

### 다양한 학습 모델 설계

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=32)

decision_tree.fit(x_train, y_train)

# RandomForest
from sklearn.ensemble import RandomForestClassifier

random_forest=RandomForestClassifier(random_state=32)

random_forest.fit(x_train, y_train)

# SVM
from sklearn import svm

svm_model=svm.SVC()

svm_model.fit(x_train, y_train)

# SGD
from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier()

sgd_model.fit(x_train, y_train)

#Logistic Regression
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(max_iter=10000)

logistic_model.fit(x_train, y_train)
```

### 다양한 학습 모델 평가

```python
from sklearn.metrics import accuracy_score

# Decision Tree
dt_y_pred = decision_tree.predict(x_test)
dt_accuracy = accuracy_score(y_test, dt_y_pred)
dt_report = classification_report(y_test, dt_y_pred)
print("Decision Tree:", dt_accuracy)
print(dt_report)


# RandomForest
rf_y_pred = random_forest.predict(x_test)
rf_accuracy = accuracy_score(y_test, rf_y_pred)
rf_report = classification_report(y_test, rf_y_pred)
print("Random Forest:", rf_accuracy)
print(rf_report)

# SVM
svm_y_pred = svm_model.predict(x_test)
svm_accuracy = accuracy_score(y_test, svm_y_pred)
svm_report = classification_report(y_test, svm_y_pred)
print("SVM:", svm_accuracy)
print(svm_report)

# SGD
sgd_y_pred = sgd_model.predict(x_test)
sgd_accuracy = accuracy_score(y_test, sgd_y_pred)
sgd_report = classification_report(y_test, sgd_y_pred)
print("SGD:", sgd_accuracy)
print(sgd_report)

# Logistic Regression
logistic_y_pred = logistic_model.predict(x_test)
logistic_accuracy = accuracy_score(y_test, logistic_y_pred)
logistic_report = classification_report(y_test, logistic_y_pred)
print("Logistic:", logistic_accuracy)
print(logistic_report)
```

유방암 진단의 경우, 가장 중요한 것은 실제 암환자를 암환자로 판별하는 것이다.
이 모델의 경우 실제 유방암 환자, 즉 label이 0인 데이터를 잘못 예측하는 경우가 적어야 하므로 정확도 평가에서 0에 대한 recall값이 가장 큰 모델이 성능이 좋다고 할 수 있다.

위에서 0에 대한 recall값은 logistic, Random Forest가 95로 가장 크며 그 중 accuracy가 가장 높은 Random Forest가 성능이 제일 좋다.
